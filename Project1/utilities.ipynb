# Convert webp to jpg

```python
from PIL import Image
import os

def convert_webp_to_jpg(folder_path):
    # Check if folder path exists
    if not os.path.exists(folder_path):
        print(f"The folder {folder_path} does not exist.")
        return

    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.webp'):
            webp_path = os.path.join(folder_path, filename)
            jpg_path = os.path.join(folder_path, filename.replace('.webp', '.jpg'))

            # Open the .webp image and convert to .jpg
            with Image.open(webp_path) as img:
                img = img.convert('RGB')  # Convert image to RGB mode
                img.save(jpg_path, 'JPEG')
                print(f"Converted {filename} to {jpg_path}")

# Example usage:
folder_path = '/Users/fredmac/Documents/DTU-FredMac/AI-generated hotdog images'
convert_webp_to_jpg(folder_path)
```

```python
import os

def delete_webp_files(folder_path):
    # Check if the folder path exists
    if not os.path.exists(folder_path):
        print(f"The folder {folder_path} does not exist.")
        return

    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.webp'):
            webp_path = os.path.join(folder_path, filename)
            os.remove(webp_path)
            print(f"Deleted {filename}")

# Example usage:
folder_path = '/Users/fredmac/Documents/DTU-FredMac/AI-generated hotdog images'
delete_webp_files(folder_path)
```

# Visualize table

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
optimizers = ['Adam', 'Adam', 'Adam', 'Adam', 
              'SGD', 'SGD', 'SGD', 'SGD', 
              'SGD Momentum', 'SGD Momentum', 'SGD Momentum', 'SGD Momentum']
lrs = ['1e-3', '1e-4', '1e-3', '1e-4', 
       '1e-2', '1e-3', '1e-2', '1e-3', 
       '1e-2', '1e-3', '1e-2', '1e-3']
batchnorm = ['False', 'False', 'True', 'True', 
             'False', 'False', 'True', 'True', 
             'False', 'False', 'True', 'True']

train_acc = [85.22, 77.70, 100, 100, 
             53.27, 47.53, 99.94, 88.09, 
             77.40, 51.19, 100, 99.94]
test_acc = [76.26, 71.91, 73.85, 75.67, 
            48.55, 51.93, 75.93, 76.26, 
            73.15, 48.07, 76.53, 76.26]

# Split data by optimizer
adam_indices = [i for i in range(len(optimizers)) if optimizers[i] == 'Adam']
sgd_indices = [i for i in range(len(optimizers)) if optimizers[i] == 'SGD']
sgd_momentum_indices = [i for i in range(len(optimizers)) if optimizers[i] == 'SGD Momentum']

# Function to format xtick labels
def format_label(lr, bn):
    return f'lr={lr}, BN' if bn == 'True' else f'lr={lr}'

# Create subplots for each optimizer
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Bar width for all subplots
bar_width = 0.35

# Adam
ax = axes[0]
index = np.arange(len(adam_indices))
train_bar = ax.bar(index - bar_width/2, [train_acc[i] for i in adam_indices], bar_width, label='Train Accuracy')
test_bar = ax.bar(index + bar_width/2, [test_acc[i] for i in adam_indices], bar_width, label='Test Accuracy')
ax.set_title('Adam', fontsize=16)  # Larger title
ax.set_xticks(index)
ax.set_xticklabels([format_label(lrs[i], batchnorm[i]) for i in adam_indices], rotation=45, ha="right", fontsize=16)  # Larger xtick labels

# SGD
ax = axes[1]
index = np.arange(len(sgd_indices))
train_bar = ax.bar(index - bar_width/2, [train_acc[i] for i in sgd_indices], bar_width, label='Train Accuracy')
test_bar = ax.bar(index + bar_width/2, [test_acc[i] for i in sgd_indices], bar_width, label='Test Accuracy')
ax.set_title('SGD', fontsize=16)  # Larger title
ax.set_xticks(index)
ax.set_xticklabels([format_label(lrs[i], batchnorm[i]) for i in sgd_indices], rotation=45, ha="right", fontsize=16)  # Larger xtick labels

# SGD Momentum
ax = axes[2]
index = np.arange(len(sgd_momentum_indices))
train_bar = ax.bar(index - bar_width/2, [train_acc[i] for i in sgd_momentum_indices], bar_width, label='Train Accuracy')
test_bar = ax.bar(index + bar_width/2, [test_acc[i] for i in sgd_momentum_indices], bar_width, label='Test Accuracy')
ax.set_title('SGD Momentum', fontsize=16)  # Larger title
ax.set_xticks(index)
ax.set_xticklabels([format_label(lrs[i], batchnorm[i]) for i in sgd_momentum_indices], rotation=45, ha="right", fontsize=16)  # Larger xtick labels

# Shared settings for all subplots
axes[0].set_ylabel('Accuracy (%)', fontsize=14)  # Larger y-axis label
fig.legend(['Train Accuracy', 'Test Accuracy'], loc='upper center', ncol=2, fontsize=12)  # Larger legend

# Display the plot
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('hyperparams.png', dpi=500)
plt.show()
```

```python
import os

def rename_jpg_files(folder_path):
    # Get a list of all jpg files in the folder
    jpg_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.jpg')]

    # Sort files to ensure they are renamed in a specific order (optional)
    jpg_files.sort()

    # Loop through the files and rename them
    for i, filename in enumerate(jpg_files, start=1):
        # Construct the new file name
        new_filename = f"AI Image {i}.jpg"

        # Create full file paths
        old_file = os.path.join(folder_path, filename)
        new_file = os.path.join(folder_path, new_filename)

        # Rename the file
        os.rename(old_file, new_file)

        print(f"Renamed: {filename} -> {new_filename}")

# Example usage
folder_path = '/Users/fredmac/Documents/DTU-FredMac/AI-generated hotdog images'
rename_jpg_files(folder_path)
```
